```python
# perform fetch / process for all regions
from tqdm import tqdm

for i in tqdm(range(490, len(moroccan_regions))):
    # read region properties
    lat = moroccan_regions["lat"][i]
    lon = moroccan_regions["lon"][i]
    region = moroccan_regions["region"][i]
    name = moroccan_regions["name"][i]

    raw_data = LocationWeather(lat, lon).weather_data_frame
    processed_data = process_data(raw_data)

    path = f"raw_data/weather_regions/{lat};{lon};{region};{name}.csv"
    processed_data.to_csv(path)
```
