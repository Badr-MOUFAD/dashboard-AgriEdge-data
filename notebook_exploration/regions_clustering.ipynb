```python
# imports
import numpy as np
import pandas as pd
# imports for plots
from plotly import graph_objects as go
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.io as pio

pio.templates.default = "plotly_white"
```

```python
df_multi_clustering = pd.DataFrame()

dir_name = "processed_data/individual_regions/"
for file in os.listdir(dir_name):
    # region lat and lon
    lat, lon, _, _ = file[:-4].split(";")

    # info region
    dict_region = np.load(dir_name + file, allow_pickle='TRUE').item()

    nb_clusters_multi = dict_region["multivariate"]["nb_clusters"]
    # weather_var, nb_cluster_uni, _ = max([(key, val['nb_clusters'], val["score"]) for key, val in dict_region["univariate"].items()], key=lambda x: x[2])

    dict_row = {"lat": [lat], "lon": [lon], "multi_nb_clusters": [nb_clusters_multi]}

    for weather_var in dict_region["univariate"]:
        dict_row[f"{weather_var}_score"] = [dict_region["univariate"][weather_var]["score"]]
        dict_row[f"{weather_var}_nb_clusters"] = [dict_region["univariate"][weather_var]["nb_clusters"]]

    df_multi_clustering = df_multi_clustering.append(pd.DataFrame(dict_row), ignore_index=True)
```

```python
# save data
df_multi_clustering.to_csv("processed_data/summary_clustering.csv", index=False)
```
