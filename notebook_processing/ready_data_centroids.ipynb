```python
data_region = pd.read_csv("raw_data/weather_regions/21.79102828115297;-14.47179828879979;Western Sahara;Western Sahara.csv")
```

```python
# multi varaite clustering
from py_scripts.clustering import multivariate_clustering

dict_multivariate = multivariate_clustering(data_region)
```

```python
dict_multivariate
```

```python
# general params
selected_cols = ["cumulative_GDD", "cumulative_PRECTOT", "cumulative_RH2M", "cumulative_WS2M"]
```

```python
# json structure
{ 
    "lat": 0,
    "lon": 0,
    "region": "region",
    "name": "",
    "cumulative_PRECTOT": {
        "High": [],
        "Medium": [],
        "Low": []
    },
    "cumulative_GDD": {
        "High": [],
        "Medium": [],
        "Low": []
    },
    "cumulative_RH2M": {
        "High": [],
        "Medium": [],
        "Low": []
    },
    "cumulative_WS2M": {
        "High": [],
        "Medium": [],
        "Low": []
    }
}
```

```python
dict_centroids = { 
    "lat": 0,
    "lon": 0,
    "region": "",
    "name": "",
    "cumulative_PRECTOT": {},
    "cumulative_GDD": {},
    "cumulative_RH2M": {},
    "cumulative_WS2M": {}
}

# loop over cluster
for cluster_name, arr_crop_years in dict_multivariate.items():
    # loop over weather variables
    for col in selected_cols:
        arr_records = []

        # loop over years of a cluster
        # append record of weather var in the given year
        for year in arr_crop_years:
            arr_records.append(
                data_region[data_region["crop_year"] == year][col]
            )

        # store the centroid
        dict_centroids[col][cluster_name] = list(np.mean(arr_records, axis=0))
```
