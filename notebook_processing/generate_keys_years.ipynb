```python
# load json
import json

with open("ready_to_use_data/year_clusters/multi_clustering.json") as f:
    multi_clustering = json.load(f)

with open("ready_to_use_data/year_clusters/uni_clustering.json") as f:
    uni_clustering = json.load(f)
```

```python
# save sub jsons
import requests


URL = "https://api.jsonstorage.net/v1/json"

dict_ids_uni = {}
dict_ids_multi = {}

# multi
filename_template = "ready_to_use_data/years_multi"
for year, data_year in multi_clustering.items():
    # save
    with open(f"{filename_template}/{year}.json", "w") as f:
        json.dump(data_year, f, indent=4)

    # push json to url and retreive id
    response = requests.post(URL, json=data_year)
    id_json = response.json()["uri"].split("/")[-1]
    dict_ids_multi[year] = id_json

# uni
filename_template = "ready_to_use_data/years_uni"
for year, data_year in uni_clustering.items():
    # save
    with open(f"{filename_template}/{year}.json", "w") as f:
        json.dump(data_year, f, indent=4)

    # push json to url and retreive id
    response = requests.post(URL, json=data_year)
    id_json = response.json()["uri"].split("/")[-1]
    dict_ids_uni[year] = id_json
```
