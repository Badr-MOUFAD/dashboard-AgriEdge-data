```python
# store ids of centroids data
from tqdm import tqdm
import requests
import json


URL = "https://api.jsonstorage.net/v1/json"
dict_ids = {}

dir_name = "ready_to_use_data/centroids/"
for file in tqdm(os.listdir(dir_name)):
    # key of regions
    key = file[:-5]
    
    # load json of the given region
    with open(f"{dir_name}/{file}") as f:
        dict_centroids = json.load(f)

    # push json to url and retreive id
    response = requests.post(URL, json=dict_centroids)
    id_json = response.json()["uri"].split("/")[-1]

    # save id
    dict_ids[key] = id_json
```
